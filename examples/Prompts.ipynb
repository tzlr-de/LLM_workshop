# 1. Ollama

````python
# Get list of models
curl http://0.0.0.0:11435/api/tags | jq ".models[].name"

# generate answer to a question
curl -X POST http://0.0.0.0:11435/api/generate -d '{"model": "gemma3:12b","prompt": "Welche Sehenswürdigkeiten gibt es in Erfurt?", "stream": false}' | jq ".response"

# ollama befehle
docker exec -it ollama ollama list
docker exec -it ollama ollama ps
docker exec -it ollama ollama run gemma3:12b "Welche Sehenswürdigkeiten gibt es in Erfurt?"

# Pull a small model
docker exec -it ollama ollama pull gemma3:1b
````
